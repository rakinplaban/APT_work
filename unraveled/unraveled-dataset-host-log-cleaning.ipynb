```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Base directory where your dataset is stored
base_directory = "D:\\DU files\\APT\\focused\\unraveled\\unraveled APT\\data\\network-flows\\cleaned"

# Recursively explore all directories and list CSV files
def explore_dataset(base_dir):
    data_info = []
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                file_size = os.path.getsize(file_path)
                data_info.append({
                    'directory': root,
                    'file': file,
                    'size': file_size,
                })
    return pd.DataFrame(data_info)

# Get dataset information
dataset_info = explore_dataset(base_directory)

# Display the dataset structure
print(dataset_info.head(10))
```

```python
import os
import pandas as pd

# Base directory where your dataset is stored
base_directory = "D:\\DU files\\APT\\focused\\unraveled\\unraveled APT\\data\\network-flows\\cleaned"

# Exclude Week 1 and list CSV files from Weeks 2 to 6
csv_files = []
for root, dirs, files in os.walk(base_directory):
    # Exclude Week 1 directories
    if "Week1" not in root:
        for file in files:
            if file.endswith('.csv'):
                csv_files.append(os.path.join(root, file))

# Load all CSV files from relevant weeks
dataframes = [pd.read_csv(file) for file in csv_files]
combined_data = pd.concat(dataframes, ignore_index=True)

# Display the first few rows of the combined data
print(combined_data.head())

# Check label distribution for balance
signature_counts = combined_data['Signature'].value_counts()
print("Signature distribution:")
print(signature_counts)

activity_counts = combined_data['Activity'].value_counts()
print("Activity distribution:")
print(activity_counts)

stage_counts = combined_data['Stage'].value_counts()
print("Stage distribution:")
print(stage_counts)
```

```python
import os
import pandas as pd
import warnings

# Ensure warnings are displayed
warnings.filterwarnings("always", category=pd.errors.DtypeWarning)

# Base directory where your dataset is stored
base_directory = "D:\\DU files\\APT\\focused\\unraveled\\unraveled APT\\data\\network-flows\\cleaned"

# List all CSV files in the base directory and its subdirectories
csv_files = []
for root, _, files in os.walk(base_directory):
    for file in files:
        if file.endswith('.csv'):
            csv_files.append(os.path.join(root, file))

# Check each CSV file individually for DtypeWarning
problematic_files = []

for file_path in csv_files:
    try:
        # Load the CSV file with low_memory=True to trigger DtypeWarning
        pd.read_csv(file_path, low_memory=True)
    except pd.errors.DtypeWarning as e:
        print(f"DtypeWarning encountered with file: {file_path}")
        problematic_files.append((file_path, str(e)))

# If there are problematic files, list them
if problematic_files:
    print("Problematic CSV files:")
    for file, warning in problematic_files:
        print(f"{file}: {warning}")
else:
    print("No DtypeWarnings encountered.")
```
